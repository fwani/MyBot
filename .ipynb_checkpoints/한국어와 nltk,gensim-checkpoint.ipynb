```python
import codecs

def read_data(filename):
    with codecs.open(filename, 'r','utf-8') as f:
        data = [line.split('\t') for line in f.read().splitlines()]
        data = data[1:]
    return data

train_data = read_data('nsmc-master/ratings_train.txt')
test_data = read_data('nsmc-master/ratings_test.txt')
```

```python
# row, column의 수가 제대로 읽혔는지 확인
print(len(train_data))      # nrows: 150000
print(len(train_data[0]))   # ncols: 3
print(len(test_data))       # nrows: 50000
print(len(test_data[0]))     # ncols: 3
```

```python
from konlpy.tag import Twitter
pos_tagger = Twitter()

def tokenize(doc):
    # norm, stem 은 optional
    return ['/'.join(t) for t in pos_tagger.pos(doc, norm=True, stem=True)]

train_docs = [(tokenize(row[1]), row[2]) for row in train_data]
test_docs = [(tokenize(row[1]), row[2]) for row in test_data]

# 잘 들어갔는지 확인
from pprint import pprint
pprint(train_docs[0])
```

```python
print(train_docs[0])
```

```python
with codecs.open("train_docs.voc","w","utf-8") as f:
    for w in train_docs:
        f.write('\t'.join(map(str, w)) + "\t")
with codecs.open("test_docs.voc","w","utf-8") as f:
    for w in test_docs:
        f.write('\t'.join(map(str, w)) + "\t")
```

```python
#Training data의 token 모으기
tokens = [t for d in train_docs for t in d[0]]
print(len(tokens))
```

```python
import nltk
text = nltk.Text(tokens, name='NMSC')
print(text)
```

```python
print(len(text.tokens))                 # returns number of tokens
print(len(set(text.tokens)))            # returns number of unique tokens
pprint(text.vocab().most_common(10))    # returns frequency distribution
```

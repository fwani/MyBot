```python
data = ''
with open("TheTerrorLive.txt", "r") as fp:
    data = fp.read()
print(data)
```

```python
import konlpy
import nltk

# POS tag a sentence
words = konlpy.tag.Twitter().pos(data)

# Define a chunk grammar, or chunking rules, then chunk
grammar = """
NP: {<N.*>*<Suffix>?}   # Noun phrase
VP: {<V.*>*}            # Verb phrase
AP: {<A.*>*}            # Adjective phrase
"""
parser = nltk.RegexpParser(grammar)
chunks = parser.parse(words)
for subtree in chunks.subtrees():
    if subtree.label()=='NP':
        print(' '.join((e[0] for e in list(subtree))))
        print(subtree.pprint())
```

```python
from konlpy.corpus import kobill    # Docs from pokr.kr/bill
files_ko = kobill.fileids()         # Get file ids
doc_ko = open('TheTerrorLive.txt').read()

from konlpy.tag import Twitter; t = Twitter()
tokens_ko = t.morphs(doc_ko)
print(tokens_ko)

import nltk
ko = nltk.Text(tokens_ko, name='TheTerrorLive')   # For Python 2, input `name` as u'유니코드'
```

```python
ko.similar('라디오')
ko.count('라디오')
```

```python
ko.plot(50)
```

```python
import re
p = re.compile('[^\u3131-\u3163\uac00-\ud7a3]+')
with open('TTL_word.word','wb') as f:
    for word in ko:
        w = p.sub('',word)
        if w == '':
            continue
        f.write((word+"\n").encode('utf-8'))
```
